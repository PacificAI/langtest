Whether you are using **John Snow Labs, Hugging Face, Spacy** models or **OpenAI, Cohere, AI21, Hugging Face Inference API and Azure-OpenAI** based LLMs, it has got you covered. You can test any Named Entity Recognition (NER), Text Classification, fill-mask, Translation model using the library. We also support testing LLMS for Question-Answering, Summarization and text-generation tasks on benchmark datasets. The library supports 100+ out of the box tests. For a complete list of supported test categories, please refer to the [documentation](http://langtest.org/docs/pages/docs/test_categories).

# Harness and Its Parameters

The Harness class is a testing class for Natural Language Processing (NLP) and LLM models. It evaluates the performance of a NLP model on a given task using test data and generates a report with test results.Harness can be imported from the LangTest library in the following way.

```python
#Import Harness from the LangTest library
from langtest import Harness
```

### Set environment for OpenAI

```python
import os

os.environ["OPENAI_API_KEY"] = "<YOUR_API_KEY>"
```

## AMEGA

```python
harness = Harness(
    task="question-answering",
    model={
        "model": "gpt-4o-mini",
        "hub": "openai",
        "type": "chat"
    },
    data=[],
    config={
        "tests": {
            "defaults": {
                "min_pass_rate": 0.8
            },
            "clinical": {
                "amega" : {
                    "no_of_cases": 5,
                }
            }
        }
    }
)
```

### Generating the test cases.

```python
harness.generate()
```

harness.generate() method automatically generates the test cases (based on the provided configuration)

### Running the tests

```python
harness.run()
```

Called after harness.generate() and is to used to run all the tests.  Returns a pass/fail flag for each test.

### Final Results

We can call `.report()` which summarizes the results giving information about pass and fail counts and overall test pass/fail flag.

```python
harness.report()
```
